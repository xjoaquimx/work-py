```python
driver = webdriver.Chrome()
```

```python

driver.get('http://10.101.50.115/sws.application/home/homeDeviceInfo.sws')
```

```python
def ValidaOn():
    try:
        driver.find_element(By.XPATH, '//*[@id="swshomeTitle"]/tbody/tr/td[1]/span')
        return True
    except:
        return False
```

```python
def DadosGerados(dados_printer, ip):
    dicionario = {}
    if ValidaOn():
        dados_prontos = []
            
        dados_printer_1 = []
        for data in dados_printer:
            for d in data:
                dados_printer_1.append(d.text.strip())
                
        for dado in dados_printer_1:
            if dado.split('\n') != ['']:
                dado = dado.split('\n')
            for d in dado:
                if d.split('\t') != ['']:
                    d = d.split('\t')
                for n in d:
                    if n.split(' ') != ['']:
                        n = n.split(' ')
                        dados_prontos.append(" ".join(n))
    
        for i in range(0, len(dados_prontos[:-15]), 2):
            if i == 0:
                dicionario['IP'] = ip
                dicionario[dados_prontos[i]] = dados_prontos[i + 1]
            dicionario[dados_prontos[i]] = dados_prontos[i + 1]
        return dicionario
    else:
        dicionario['IP'] = ip
        dicionario['Estado'] = 'OFF'
        return dicionario
```
